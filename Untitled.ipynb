```python
from ultralytics import YOLO
import cv2

model=YOLO("yolov8n.pt")

#opening the video file path
#0 for live camera capturing and you can also give the link of video file here

video_path="video.mp4"
cap=cv2.VideoCapture(video_path)

# loop through the video frames
while cap.isOpened():
    
    # reading the frame from the video
    success,frame=cap.read()
    
    if success:
        # running yolo model on frames
        new_results=model.track(frame,persist=True)
        
        # visualize the result on the frame
        annotated_frame=new_results[0].plot()
        
        # display the annotated frame
        cv2.imshow("YOLO annotations",annotated_frame)
        
        # breaking the loop by pressing key
        if cv2.waitKey(1)==13: #13 for Enter key
            break
    else:
        # break the loop if video ended
        break
cap.release()
cv2.destroyAllWindows()
```
